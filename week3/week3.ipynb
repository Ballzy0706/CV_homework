```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"D:\CV_JianZhang\week3\Lena.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
harrface = cv2.CascadeClassifier(r"D:\CV_JianZhang\week3\trainer\haarcascades\haarcascade_frontalface_alt2.xml")
def show_detection(image, faces):
    """在每个检测到的人脸上绘制一个矩形进行标示"""
    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 5)
    return image
def show_img_with_matplotlib(color_img, title, pos):
    img_RGB = color_img[:, :, ::-1]
    ax = plt.subplot(2, 2, pos)
    plt.imshow(img_RGB)
    plt.title(title, fontsize=8)
    plt.axis('off')
    plt.savefig('D:\CV_JianZhang\week3\detact_Lena.png',dpi=500, bbox_inches='tight')
```

```python
# 调用 show_detection() 函数标示检测到的人脸
faces_detact = harrface.detectMultiScale(gray)
img_faces = show_detection(img.copy(), faces_detact)
cv2.putText(img_faces, 'Lena', (100, 100), cv2.FONT_HERSHEY_TRIPLEX, 2, (0, 0, 255), 5, 16,False)
show_img_with_matplotlib(img_faces, "face: " + str(len(faces_detact)), 1)
```
